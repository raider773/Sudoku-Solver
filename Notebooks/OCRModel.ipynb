```python
import cv2 as cv
import matplotlib.pyplot as plt
from PIL import Image
import easyocr
```

```python
reader = easyocr.Reader(['en','en']) # this needs to run only once to load the model into memory
```

```python
import pickle
with open('digits.pkl', 'rb') as f:
    pruebadefuego = pickle.load(f)  
```

```python
for i in range (81):
    up_width = 400
    up_height = 400
    up_points = (up_width, up_height)
    resized_up = cv.resize(pruebadefuego[i], up_points, interpolation= cv.INTER_LINEAR)
    cv.imwrite("ActualPicture.JPG",resized_up)
    result = reader.readtext('ActualPicture.JPG')
    if len(result)> 0:
        print(result[0][1])
    else:
        print(0)
    plt.imshow(pruebadefuego[i])
    plt.show()
```
